```python
# 2002
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
output_notebook()
data2002 = pd.read_csv('C:\Users\khamz\Downloads\National Assembly 2002 - Updated.csv')
data1 = data2002[data2002.Constituency_title=='NA-1']
data2 = data2002[data2002.Constituency_title=='NA-2']
data3 = data2002[data2002.Constituency_title=='NA-3']
na1_2002 = data1['Votes'].sum()
na2_2002 = data2['Votes'].sum()
na3_2002 = data3['Votes'].sum()
```

```python
# 2008
import pandas as pd
data2008 = pd.read_csv('C:\Users\khamz\Downloads\National Assembly 2008.csv')
data4 = data2008[data2008.ConstituencyTitle =='NA-1']
data5 = data2008[data2008.ConstituencyTitle =='NA-2']
data6 = data2008[data2008.ConstituencyTitle =='NA-3']
na1_2008 = data4['Votes'].sum()
na2_2008 = data5['Votes'].sum()
na3_2008 = data6['Votes'].sum()
```

```python

import pandas as pd
data2013 = pd.read_csv('C:\Users\khamz\Downloads\National Assembly 2013 - Updated.csv')
data7 = data2008[data2013.ConstituencyTitle =='NA-1']
data8 = data2008[data2013.ConstituencyTitle =='NA-2']
data9 = data2008[data2013.ConstituencyTitle =='NA-3']
na1_2013 = data7['Votes'].sum()
na2_2013 = data8['Votes'].sum()
na3_2013 = data9['Votes'].sum()
```

```python
from bokeh.palettes import GnBu3, OrRd3

NAs = ['NA-1', 'NA-2', 'NA-3']
years = ['2002', '2008', '2013']
votes = {'NAs' : NAs,
           '2002' :[na1_2002, na2_2002, na3_2002],
                                   
           '2008' :[na1_2008, na2_2008, na3_2008],
                                              
           '2013' :[na1_2013, na2_2013, na3_2013]}

p = figure(y_range=NAs, plot_height=500, title="2002, 2008, and 2013 NA Votes in Pakistan")

p.hbar_stack(years, y='NAs', height=0.9, color=OrRd3, source=ColumnDataSource(votes),
             legend=["%s votes" % x for x in years])

p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.legend.location = "top_right"

show(p)
```
